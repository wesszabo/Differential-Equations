<div class='exercise'>
Tegyük fel, hogy a korábbi sóoldatos feladatban a befolyó sómennyiség exponenciálisan csökken.
Ekkor az egyenlet

\begin{equation}\label{eq:origin}
y'(t)=0.6e^{-t}-0.2y(t).
\end{equation}

Adjuk meg az általános megoldást.
</div>

<div class='solution'>

(a) Először a homogén egyenletet oldjuk meg. 

$$
y'(t)=-0.2y(t).
$$

Az egyenlet konstans függvény megoldása

$$
y(t)=\underset{\sim}{0}.
$$
</div>

<div class='solution_no_word'>
Az egyenletet

$$
\frac{dy}{dt}=-0.2y
$$

alakban írva a változókat szét tudjuk választani

$$
\frac{1}{y}\,dy=-0.2\,dt.
$$

Mindkét oldalt integrálva kapjuk, hogy

$$
\ln|y|=-0.2t+C,
$$

ahol $C$
</div>

<div class='solution_no_word'>
tetszőleges konstans.
</div>

<div class='solution_no_word'>
Rendezve az egyenletet, 

$$
|y|=C e^{-0.2t},
$$

ahol $C$
</div>

<div class='solution_no_word'>
tetszőleges pozitív konstans.
</div>

<div class='solution_no_word'>
Ebből 

$$
y(t)=Ce^{-0.2t},
$$

ahol $C$
</div>

<div class='solution_no_word'>
tetszőleges nem-nulla konstans. 
</div>

<div class='solution_no_word'>
A $C=0$ esetben megkapjuk a konstans függvény megoldást. 
Ezért a homogén egyenlet összes megoldása

$$
y_{hö}(t)=Ce^{-0.2t},
$$

ahol $C$ tetszőleges valós szám lehet.
</div>

<div class='solution_no_word'>
(b) Az inhomogén egyenlet egy partikuláris megoldását 

$$
y(t)= ?
$$

alakban keressük, ahol
</div>

<div class='solution_no_word'>
$$
y(t)=c(t)e^{-0.2t}.
$$
</div>

<div class='solution_no_word'>
Ebből 
$$
y'(t)=c'(t)e^{-0.2t}-0.2c(t)e^{-0.2t}.
$$

Visszahelyettesítve az eredeti $\eqref{eq:origin}$  egyenletbe azt kapjuk, hogy 

\begin{align}
c'(t)e^{-0.2t} &-0.2c(t)e^{-0.2t}\\
    &=0.6e^{-t}-0.2c(t)e^{-0.2t},
\end{align}

amiből rendezés után
$$
c'(t)=0.6e^{-0.8t}
$$
adódik. 
</div>

<div class='solution_no_word'>
Integrálás után
$$
c(t)=?
$$
</div>

<div class='solution_no_word'>
$$
c(t)=-\frac{3}{4}e^{-0.8t}
$$

adódik.
</div>

<div class='solution_no_word'>
Tehát az inhomogén egyenlet egy partikuláris megoldása

$$
y_{ihp}(t)=?
$$
</div>

<div class='solution_no_word'>
$$
y_{ihp}(t)=-\frac{3}{4}e^{-t}.
$$
</div>

<div class='solution_no_word'>
Az inhomogén egyenlet összes megoldása 

$$
y_{ihö}(t)=?
$$
</div>

<div class='solution_no_word'>
$$
y_{ihö}(t)=y_{hö}(t)+y_{ihp}(t),
$$
</div>

<div class='solution_no_word'>
azaz,

$$
y_{ihö}(t)=Ce^{-0.2t}-\frac{3}{4}e^{-t},
$$

ahol $C$ tetszőleges valós szám. $\blacksquare$
</div>

<div class='remark'>
\begin{equation}\label{eq:}
\lim_{t\to\infty}y(t)=0,
\end{equation}
azaz, a megoldás az egyensúlyi megoldáshoz tart, a kezdeti értéktől függetlenül.
</div>

```python
import plotly.figure_factory as ff
import plotly.graph_objects as go
import numpy as np
import ipywidgets as widgets
from ipywidgets import interactive, VBox, FloatSlider 

C = 3
t = np.linspace(0, 10, 100)
y = C*np.exp(-0.2*t) - 3./4.*np.exp(-t)

trace1 = go.Scatter(
    line=dict(color='red', width=2),
    x=t,
    y=y,
    name=r'$y(t)$')

trace2 = go.Scatter(
    line=dict(color='black', width=1),
    x=t,
    y=0*t,
    name=r'egyensúlyi megoldás')

data = [trace1, trace2]

layout = go.Layout(
    xaxis = dict(
        title = r'time [minute]',
        range=[0,10]
    ),
    yaxis = dict(
        title = r'salt [kg]',
        range = [-5,5],
        scaleratio = 1
    ),
    title = dict(
        text = r'$y(t)=C e^{-0.2t} - \frac{3}{4}e^{-t}$',
        font = dict(
            size = 20
        ),
        y = 0.85,
        x = 0.5,
        xanchor = 'center',
        yanchor = 'top'
    )
)

fig = go.FigureWidget(
    data = data, layout = layout
)

###############
def update_y(C):
    fig.data[0].y = C*np.exp(-0.2*t) - 3./4.*np.exp(-t)

style = {'description_width': 'initial'}
slider_C =  FloatSlider(min=-3, max=5, step=0.1, value=3,
                       style=style)
slider_C.description = r'$C$'
slider = interactive(update_y, C=slider_C)
vb = VBox((fig, slider_C))
vb.layout.align_items = 'center'
vb
```
