<h1>5. gyakorlat első integrál 2.<span class="tocSkip"></span></h1>
<div class="toc"><ul class="toc-item"></ul></div>

<div class='exercise'>
Tekintsük a rugó által mozgatott test mozgásegyenletét: $m\ddot{x}+Dx=0$, ahol $m,D>0$ állandók. 
Írjuk át az egyenletet rendszerré, keressünk hozzá első integrált, majd írjuk fel az egyenletre az energiamegmaradást és ábrázoljuk a fázisképet az $x,\dot{x}$ síkon.
</div>

<div class='solution'>

\begin{cases} 
\begin{align}
\dot{x} &=y, \\[3pt]
\dot{y} &=-\frac{D}{m}x. 
\end{align}
\end{cases}
       
Az egyensúlyi  pontokat az $(\dot{x},\dot{y})=(0,0)$ egyenletből határozhatjuk meg. 
    
Kapjuk, hogy $x={0}$, $y={0}$.
</div>

<div class='solution_no_word'>

A további megoldásokhoz válasszuk szét a változókat.

\begin{cases} 
\begin{align}
\frac{1}{y}\,dx &=1\,dt \\[4pt]
-\frac{m}{D}\cdot\frac{1}{x}\,dy &=1\,dt. 
\end{align}
\end{cases}

A jobb oldalak egyenlőségéből következik, hogy 

$$
\frac{1}{y}\,dx = -\frac{m}{D}\cdot\frac{1}{x}\,dy,
$$

azaz,

$$
x\,dx +\frac{m}{D}y\,dy=0.
$$
</div>

<div class='solution_no_word'>
Mivel $(x)'_y=0=\left(\dfrac{m}{D}y\right)'_x$, ezért az egyenlet egzakt. 

Tehát létezik olyan $F(x,y)$, hogy 

\begin{cases} 
\begin{align}
F'_x(x,y) &=x, \\[3pt]
F'_y(x,y) &=\frac{m}{D}y. 
\end{align}
\end{cases}

Az első egyenletből következik, hogy 

$$
F(x,y)=\frac{x^2}{2}+C(y).
$$
</div>

<div class='solution_no_word'>
A második egyenletbe helyettesítve kapjuk, hogy

$$
\frac{m}{D}y=F'_y(x,y)=C'(y).
$$

Ebből $C(y)=\dfrac{m}{D}\dfrac{y^2}{2}$. Ezért $F(x,y)=\dfrac{x^2}{2}+\dfrac{m}{D}\dfrac{y^2}{2}$. Tehát az első integrál 

$$
x^2+\dfrac{m}{D}y^2=C,
$$

ahol $C$ tetszőleges konstans.
</div>

<div class='solution_no_word'>
Mivel $C$ független a $t$ időtől, ezért a $t=0$ időpillanatot behelyettesítve kapjuk, hogy 

$$
x^2(t)\!+\!\frac{m}{D}(\dot{x}(t))^2\!=\!x^2(0)\!+\!\frac{m}{D}(\dot{x}(0))^2.
$$
</div>

```python
import plotly.figure_factory as ff
import plotly.graph_objects as go
import numpy as np
import ipywidgets as widgets
from ipywidgets import interactive, VBox, FloatSlider 

x,y = np.meshgrid(np.arange(-5.5, 5.5, .4), np.arange(-4.5, 4.5, .4))
m = 1.
D = 2.
u = y
v = - D/m * x

angle = np.arctan(v / u)
row, col = angle.shape
for r in range(row):
    for c in range(col):
        if u[r][c] < 0:
            angle[r][c] = angle[r][c] + np.pi

norm = 0.30
u = norm * np.cos(angle)
v = norm * np.sin(angle)

quiver = ff.create_quiver(x, y, u, v, scale = 1)

x0 = 1. # = x(0)
y0 = 2. # = x'(0)
C = x0**2 + m/D * y0**2

# Add contour to figure
delta = 0.1
xrange = np.arange(-5.5, 5.5, delta)
yrange = np.arange(-4.5, 4.5, delta)
X, Y = np.meshgrid(xrange,yrange)
F = X**2
G = C - m/D * Y**2

trace0 = go.Contour(
        z = F-G,
        x = xrange,
        y = yrange,
        contours_coloring='none',
        line_width = 2,
        line_color = 'rgb(228,13,13)',
        contours=dict(
            start=0,
            end=0,
            size=2
        ),
        name = 'first integral'
    )

# Add quiver to figure
trace1 = go.Scatter(
    line=dict(color='blue', width=2),
    x=quiver.data[0].x,
    y=quiver.data[0].y,
    name=r'direction')

# Add point to figure
trace2 = go.Scatter(
    x=[x0], y=[y0],
    mode = 'markers',
    marker_size = 8,
    marker_color = 'rgb(228,13,13)',
    name = 'point')

data = [trace0, trace1, trace2]

layout = go.Layout(
    xaxis = dict(
        title = r'$x$',
        range=[-4.5, 4.5]
    ),
    yaxis = dict(
        title = r"$x'$",
        range = [-4.5, 4.5],
        scaleratio = 1
    ),
    title = dict(
        text = r'$m\ddot{x}+Dx=0\quad \text{fázisképe},\quad x^2+\dfrac{m}{D}y^2=C$',
        font = dict(
            size = 20
        ),
        y = 0.85,
        x = 0.35,
        xanchor = 'center',
        yanchor = 'top'
    )
)

fig = go.FigureWidget(
    data = data, layout = layout
)
###############
    
def update(x0, y0, m, D):
    fig.data[2].x = [x0]
    fig.data[2].y = [y0]
    C = x0**2 + m/D * y0**2
    F = X**2
    G = C - m/D * Y**2
    fig.data[0].z = F - G
    print('C: ', str("%0.2f" % C))
    
style = {'description_width': 'initial'}

slider_x0 =  FloatSlider(min=-3, max=3, step=0.1, value=1,
                       style=style)
slider_x0.description = r'$x(0)$'

slider_y0 =  FloatSlider(min=-3, max=3, step=0.1, value=2,
                       style=style)
slider_y0.description = r"$x'(0)$"

slider_m =  FloatSlider(min=0.1, max=3, step=0.1, value=2,
                       style=style)
slider_m.description = r"$m$"

slider_D =  FloatSlider(min=0.1, max=3, step=0.1, value=2,
                       style=style)
slider_D.description = r"$D$"

slider = interactive(update, x0=slider_x0, y0=slider_y0, m=slider_m, D=slider_D)

vb = VBox((fig, slider))
vb.layout.align_items = 'center'
vb
```
