<div class="toc"><ul class="toc-item"></ul></div>

<div class='exercise'>
Igazoljuk, hogy bármely $f\in C(\mathbf{R}^2)$ esetén az 
    
\begin{cases} 
\begin{align}
\dot{x} &=f(x,y), \\[3pt]
\dot{y} &=-f(x,y). 
\end{align}
\end{cases}
    
rendszernek $F(x,y):=x+y$ első integrálja. (A konkrét $f$ tehát csak azt határozza meg, 
hogy milyen gyorsan mozognak a pontok a $t$ "idő" függvényében az egyenes pályákon.)
</div>

<div class='solution'>

Az egyensúlyi  pontokat az $(\dot{x},\dot{y})=(0,0)$ egyenletből határozhatjuk meg. 
    
Kapjuk, hogy $f(x,y)=0$.
</div>

<div class='solution_no_word'>
A további megoldásokhoz válasszuk szét a változókat.

\begin{cases} 
\begin{align}
\frac{1}{f(x,y)}\,dx &=1\,dt \\[4pt]
-\frac{1}{f(x,y)}\,dy &=1\,dt. 
\end{align}
\end{cases}

A jobb oldalak egyenlőségéből következik, hogy 

$$
\frac{1}{f(x,y)}\,dx = -\frac{1}{f(x,y)}\,dy,
$$

azaz,

$$
1\,dx +1\,dy=0.
$$
</div>

<div class='solution_no_word'>
Mivel $(1)'_y=0=(1)'_x$, ezért az egyenlet egzakt. 

Tehát létezik olyan $F(x,y)$, hogy 

\begin{cases} 
\begin{align}
F'_x(x,y) &=1, \\[3pt]
F'_y(x,y) &=1. 
\end{align}
\end{cases}

Az első egyenletből következik, hogy 

$$
F(x,y)=x+C(y).
$$
</div>

<div class='solution_no_word'>
A második egyenletbe helyettesítve kapjuk, hogy

$$
1=F'_y(x,y)=C'(y).
$$

Ebből $C(y)=y$. Ezért $F(x,y)=x+y$. Tehát az első integrál 

$$
x+y=C,
$$

ahol $C$ tetszőleges konstans.
</div>

```python
import plotly.figure_factory as ff
import plotly.graph_objects as go
import numpy as np
import ipywidgets as widgets
from ipywidgets import interactive, VBox, FloatSlider 

x,y = np.meshgrid(np.arange(-5.5, 5.5, .4), np.arange(-4.5, 4.5, .4))
u = x
v = - x

angle = np.arctan(v / u)
row, col = angle.shape
for r in range(row):
    for c in range(col):
        if u[r][c] < 0:
            angle[r][c] = angle[r][c] + np.pi

norm = 0.30
u = norm * np.cos(angle)
v = norm * np.sin(angle)

quiver = ff.create_quiver(x, y, u, v, scale = 1)

x0 = 1
y0 = 2
C = x0 + y0

# Add contour to figure
delta = 0.1
xrange = np.arange(-5.5, 5.5, delta)
yrange = np.arange(-4.5, 4.5, delta)
X, Y = np.meshgrid(xrange,yrange)
F = X
G = C - Y

trace0 = go.Contour(
        z = F-G,
        x = xrange,
        y = yrange,
        contours_coloring='none',
        line_width = 2,
        line_color = 'rgb(228,13,13)',
        contours=dict(
            start=0,
            end=0,
            size=2
        ),
        name = 'first integral'
    )

# Add quiver to figure
trace1 = go.Scatter(
    line=dict(color='blue', width=2),
    x=quiver.data[0].x,
    y=quiver.data[0].y,
    name=r'direction')

# Add point to figure
trace2 = go.Scatter(
    x=[x0], y=[y0],
    mode = 'markers',
    marker_size = 8,
    marker_color = 'rgb(228,13,13)',
    name = 'point')

data = [trace0, trace1, trace2]

layout = go.Layout(
    xaxis = dict(
        title = r'$x$',
        range=[-5.5, 5.5]
    ),
    yaxis = dict(
        title = r'$y$',
        range = [-4.5, 4.5],
        scaleratio = 1
    ),
    title = dict(
        text = r'$\dot{x}=f(x,y),\quad\dot{y}=-f(x,y)\quad \text{fázisképe},\quad x+y=C$',
        font = dict(
            size = 20
        ),
        y = 0.85,
        x = 0.35,
        xanchor = 'center',
        yanchor = 'top'
    )
)

fig = go.FigureWidget(
    data = data, layout = layout
)
###############
    
def update(x0, y0):
    fig.data[2].x = [x0]
    fig.data[2].y = [y0]
    C = x0 + y0
    F = X
    G = C - Y
    fig.data[0].z = F - G
    print('C: ', str("%0.2f" % C))
    
style = {'description_width': 'initial'}

slider_x0 =  FloatSlider(min=-3, max=3, step=0.1, value=1,
                       style=style)
slider_x0.description = r'$x_0$'

slider_y0 =  FloatSlider(min=-3, max=3, step=0.1, value=2,
                       style=style)
slider_y0.description = r'$y_0$'

slider = interactive(update, x0=slider_x0, y0=slider_y0)

vb = VBox((fig, slider))
vb.layout.align_items = 'center'
vb
```
